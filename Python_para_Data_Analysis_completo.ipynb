Você notou que ao chamar essa propriedade - .shape, o código retornou um valor dentro de dois parênteses (9994, 21)?

Essa estrutura, dentro de parênteses, é chamada de Tupla!

Tupla é um tipo de estrutura de dados utilizada em Python que funciona de modo semelhante a uma lista, entretanto, com a característica principal de ser imutável. Isso significa que quando uma tupla é criada não é possível adicionar, alterar ou remover seus elementos. Geralmente, ela é utilizada para adicionar tipos diferentes de informações, porém, com a quantidade de elementos definidos.

Podemos utilizar uma tupla de dois elementos, por exemplo, para indicar a sigla do estado em uma posição e o nome dele em outra. Portanto, ela é uma boa opção quando queremos trabalhar com informações diferentes em uma mesma variável e quando queremos que esses dados não sofram alterações.

Sua característica de imutabilidade oferece segurança nas informações armazenadas. Por isso, uma das finalidades da tupla é armazenar uma sequência de dados que não será modificada em outras partes do código.

Vamos colocar isso numa função?

````python
def agrupar_valores(df, chave, frequencia):
  """
  Agrupa valores do dataframe, indicando para a frequencia especificada:
  Média das Vendas, Soma das vendas, Soma de quantidade vendida, Média do Lucro e Soma do Lucro
  """

  agrupamento = df.groupby(pd.Grouper(key = chave, freq = frequencia)).agg({
    'Sales' : ['mean', 'sum'],
    'Quantity': 'sum',
    'Profit': ['mean', 'sum']
  })
  
  agrupamento.columns = agrupamento.columns.droplevel()

  agrupamento.columns = ['Sales_mean', 'Sales_sum',
                       'Quantity_sum',
                       'Profit_mean', 'Profity_sum']
                    
  agrupamento.reset_index(inplace = True)

  agrupamento['AUMENTO_VENDAS'] = agrupamento['Sales_sum'].diff()

  agrupamento['AUMENTO_VENDAS_PCT'] = agrupamento['Sales_sum'].pct_change()

  return agrupamento
````

**Exercício**

Fazer a mesma função, adicionando um parâmetro de Categoria

**Criar função com média ponderada**

Vamos usar de exemplo a quantidade como fator da média ponderada.


````python
def agregar_com_mp(df, chave, freq, volume):


  """
  Agrega a base real -hana diariamente.
  """
  mp = lambda x: np.ma.average(x, weights=df.loc[x.index, volume])
  mp.__name__ = 'mp'
  max_minus_min = lambda x: pd.Series(x).max()-pd.Series(x).min()
  max_minus_min.__name__ = 'max_minus_min'

  agrupamento = df.groupby(pd.Grouper(key = chave, freq = freq)).agg({
          'Sales' : ['mean', 'sum', mp, max_minus_min],
          'Quantity': 'sum',
          'Profit': ['mean', 'sum']
          })

  agrupamento.columns = agrupamento.columns.droplevel()

  agrupamento.columns = ['Sales_mean', 'Sales_sum', 'Sales_mp', 'Sales_range',
                         'Quantity_sum',
                         'Profit_mean', 'Profity_sum']
                    
  agrupamento.reset_index(inplace = True)

  agrupamento['AUMENTO_VENDAS'] = agrupamento['Sales_sum'].diff()

  agrupamento['AUMENTO_VENDAS_PCT'] = agrupamento['Sales_sum'].pct_change()

  return agrupamento
````

````python
agregar_com_mp(df, 'Order Date', 'D', 'Quantity')
````

````python
import xlsxwriter
from datetime import datetime
````

````python
def create_xlsx(vendas_dia, vendas_mes):

    dia = datetime.today().strftime('%Y-%m-%d')

    nome_arquivo = 'compilado-' + str(dia) + '.xlsx'

    writer = pd.ExcelWriter(nome_arquivo, engine= 'xlsxwriter')

    vendas_dia.to_excel(writer, sheet_name = 'vendas_dia')
    vendas_mes.to_excel(writer, sheet_name = 'vendas_mes')

    writer.save()
````
